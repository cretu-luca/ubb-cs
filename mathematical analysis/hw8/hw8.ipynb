```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def valid(x1, x2, p):
    if(abs(x1) ** p + abs(x2) ** p) ** (1 / p) < 1:
        return True
    
    return False
```

```python
def main(p):
    sample = np.linspace(-1, 1, 100)
    x_list = list()
    y_list = list()

    for x in sample:
        for y in sample:
            if valid(x, y, p):
                x_list.append(x)
                y_list.append(y)

    plt.grid()
    sizes = [7] * len(x_list)
    plt.scatter(x_list, y_list, sizes, marker='.', color='royalblue')
    plt.text(0.35, 0.35, f"p = {p}", size=14, backgroundcolor='yellow')
    plt.title(f"The unit ball with p-norm")
    plt.show() 
```

```python
main(1.25)
```

```python
main(1.5)
```

```python
main(3)
```

```python
main(8)
```
