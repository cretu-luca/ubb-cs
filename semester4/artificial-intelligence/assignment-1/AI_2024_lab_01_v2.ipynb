***Tensor Shapes***

```python
import torch

x = torch.empty(2, 2, 3)
print(x.shape)
print(f'Empty {x}')

empty_like_x = torch.empty_like(x)
print(empty_like_x.shape)
print(f'empty_like {empty_like_x}')

zeros_like_x = torch.zeros_like(x)
print(zeros_like_x.shape)
print(f'zeros_like {zeros_like_x}')

ones_like_x = torch.ones_like(x)
print(ones_like_x.shape)
print(f'ones_like {x}')

rand_like_x = torch.rand_like(x)
print(rand_like_x.shape)
print(f'rand_like {x}')
```

Conside an unfair probability spinner with $n$ slots. Each slot has a probabilty $P_n \in [0,1]$ and $\sum_{i=0}^{n-1}P_i = 1$.

### Solution

```python
import random as rand
import numpy as np
```

### Inverse transformation sampling

In order to simulate the unfair probability spinner, we will use Inverse transformation sampling, which works by: 

1. Generating a uniform random value between 0 and 1.
2. Building the cumulative distribution function (CDF) incrementally.
3. Returning an index of the first slot where the random value is less than the commulative distribution.

```python
def spinner(probabilityDistribution):
    value = rand.random() # (1.)
    cumulative_probability = 0

    for index, prob in enumerate(probabilityDistribution):
        cumulative_probability += prob # (2.)

        if value < cumulative_probability:
            return index # (3.)
        
    return index
```

```python
print(spinner([0.4, 0.1, 0.15, 0.2, 0.075, 0.05, 0.025]))
```

```python
def builtin_spinner(probabilityDistribution):
    return np.random.choice(np.arange(probabilityDistribution.size), p=probabilityDistribution)
```

```python
def simulation(spinner: callable, prob):
    count = [0] * len(prob)
    probDistribution = np.array(prob)

    for i in range(5000000):
        index = spinner(probDistribution)

        count[index] += 1
    
    for i in range(len(prob)):
        print(f"{i} - {count[i] / 5000000}")
```

```python
print("cdf:")
simulation(spinner, [0.4, 0.1, 0.15, 0.2, 0.075, 0.05, 0.025])
print("\nbuiltin:")
simulation(builtin_spinner, [0.4, 0.1, 0.15, 0.2, 0.075, 0.05, 0.025])
```

## Task 7 -- Determine some outcomes using Monte Carlo Simulations

Consider a large pipe with a radius $R = 30 cm$ that has the opening pointing straight up, as in the figure. Around this pipe is a rectangular wall with an edge $l = 1m$. Balls of radius $r = 1 cm$ fall down uniformly from above inside the square yard. Compute, using a Monte Carlo simulation, the probability that a ball falls inside the pipe. Compare the results with the mathematically deduced probability.



### Solution

```python
R = 30 # pipe radius
r = 1 # ball radius
l = 100 # wall side length
```

```python
def monte_carlo_simulation():
    trials = 100000000
    
    x = np.random.uniform(0, 100, trials)
    y = np.random.uniform(0, 100, trials)

    successes = np.count_nonzero((x - 50) ** 2 + (y - 50) ** 2 <= (R - r) ** 2)

    return successes / trials
```

```python
def theoretical_probability():
    success_area = np.pi * (R - r) ** 2
    total_area = l * l

    return success_area / total_area
```

```python
print("monte carlo simuation:")
print(monte_carlo_simulation())
print("theoretical probability:")
print(theoretical_probability())
```
